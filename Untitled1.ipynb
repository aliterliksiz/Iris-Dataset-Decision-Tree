```python
#Importing required libraries
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
#Loading the iris data
data = load_iris()
```

```python
# Extracting data attributes / features
X = data.data

# Extracting target/ class labels
y = data.target
```

```python
# Import library for splitting the dataset into train and test.
from sklearn.model_selection import train_test_split
```

```python
#Using the train_test_split to create train and test sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 50, test_size = 0.25)
```

```python
#Importing the Decision tree classifier from the sklearn library.
from sklearn.tree import DecisionTreeClassifier
#########  WHAT IS THE INFORMATIN GAIN  ###############
#Information gain is the reduction in entropy or surprise by transforming a dataset and is often used in training decision trees.
#Information gain is calculated by comparing the entropy of the dataset before and after a transformation.
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
```

```python
#prediction
y_pred = clf.predict(X_test)
```

```python
#Importing the accuracy metric from sklearn.metrics library
from sklearn.metrics import accuracy_score

##########   WHAT IS ACCURACY   ##########
```

![](accuracy.png)

```python
print('Accuracy Score on train data (using Default criterionas gini): ', accuracy_score(y_true=y_train, y_pred=clf.predict(X_train)))
print('Accuracy Score on test data (using Default criterionas gini): ', accuracy_score(y_true=y_test, y_pred=y_pred))
```

```python
#Importing the precision metric from sklearn.metrics library
from sklearn.metrics import precision_score

##########   WHAT IS PRECISION   ##########

#Compute the precision.
#The precision is the ratio tp / (tp + fp) where tp is the number of true positives and fp the number of false positives. 
#The precision is intuitively the ability of the classifier not to label as positive a sample that is negative.
```

```python
print('Precision Score on train data (using Default criterionas gini): ', precision_score(y_true=y_train, y_pred=clf.predict(X_train), average="weighted"))
print('Precision Score on test data (using Default criterionas gini): ', precision_score(y_true=y_test, y_pred=y_pred, average="weighted"))
```

```python
#Importing the recall metric from sklearn.metrics library
from sklearn.metrics import recall_score

##########   WHAT IS RECALL   ##########

#Compute the recall.
#The recall is the ratio tp / (tp + fn) where tp is the number of true positives and fn the number of false negatives. 
#The recall is intuitively the ability of the classifier to find all the positive samples.
```

```python
print('Recall Score on train data (using Default criterionas gini): ', recall_score(y_true=y_train, y_pred=clf.predict(X_train), average="weighted"))
print('Recall Score on test data (using Default criterionas gini): ', recall_score(y_true=y_test, y_pred=y_pred, average="weighted"))
```

```python
#Importing the f1 metric from sklearn.metrics library
from sklearn.metrics import f1_score

##########   WHAT IS F-1 SCORE   ##########

#Compute the F1 score, also known as balanced F-score or F-measure.
#The F1 score can be interpreted as a harmonic mean of the precision and recall, where an F1 score reaches its best value at 1 
#and worst score at 0. The relative contribution of precision and recall to the F1 score are equal. The formula for the F1 score 
#is: F1 = 2 * (precision * recall) / (precision + recall)
```

![](anothers.png)

```python
print('F-1 Score on train data (using Default criterionas gini): ', f1_score(y_true=y_train, y_pred=clf.predict(X_train), average='micro'))
print('F-1 Score on test data (using Default criterionas gini): ', f1_score(y_true=y_test, y_pred=y_pred, average='micro'))
```

```python
#Importing the confusion matrix from sklearn.metrics library
from sklearn.metrics import confusion_matrix

############   WHAT IS CONFUSION MATRIX   #############

#he confusion_matrix function evaluates classification accuracy by computing the confusion matrix with each row corresponding to 
#the true class (Wikipedia and other references may use different convention for axes).
```

```python
print('Confusion Matrix on train data (using Default criterionas gini): ', confusion_matrix(y_true=y_train, y_pred=clf.predict(X_train)))
print('Confusion Matrix on test data (using Default criterionas gini): ', confusion_matrix(y_true=y_test, y_pred=y_pred))
```

```python
###########################   ENTROPY   #############################
```

```python
clf_entropy = DecisionTreeClassifier(criterion = 'entropy')
```

```python
clf_entropy.fit(X_train, y_train)
```

```python
y_pred_entropy = clf_entropy.predict(X_test)
```

```python
print('Accuracy Score on train data (using Default criterion as Entropy): ', precision_score(y_true=y_train, y_pred =clf_entropy.predict(X_train), average="weighted"))
print('Accuracy Score on test data (using Default criterion as Entropy): ', precision_score(y_true=y_test, y_pred =y_pred_entropy, average="weighted"))
```

```python
print('Accuracy Score on train data (using Default criterion as Entropy): ', recall_score(y_true=y_train, y_pred =clf_entropy.predict(X_train), average="weighted"))
print('Accuracy Score on test data (using Default criterion as Entropy): ', recall_score(y_true=y_test, y_pred =y_pred_entropy, average="weighted"))
```

```python
print('Accuracy Score on train data (using Default criterion as Entropy): ', f1_score(y_true=y_train, y_pred =clf_entropy.predict(X_train), average='micro'))
print('Accuracy Score on test data (using Default criterion as Entropy): ', f1_score(y_true=y_test, y_pred =y_pred_entropy, average='micro'))
```

```python
print('Confusion Matrix on train data (using Default criterion as Entropy): ', confusion_matrix(y_true=y_train, y_pred=clf_entropy.predict(X_train)))
print('Confusion Matrix on test data (using Default criterion as Entropy): ', confusion_matrix(y_true=y_test, y_pred=y_pred_entropy))
```

```python
#Visualize the Tree
dot_data = StringIO()
export_graphviz(clf_entropy, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,feature_names = data.feature_names,class_names=data.target_names)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())

###########   WHAT IS ENTROPY   #########
#Entropy Another very popular way to split nodes in the decision tree is Entropy.
#Entropy is the measure of Randomness in the system.
```
